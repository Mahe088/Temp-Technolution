```python
from datetime import datetime
from datetime import timedelta
import pandas as pd
import tweepy
time_str = '10/01/2022 00:00:00.000000'
date_format_str = '%d/%m/%Y %H:%M:%S.%f'
# create datetime object from timestamp string
given_time = datetime.strptime(time_str, date_format_str)
```

```python
n=0
time_intervals = []
for i in range(289):
    n=n+1
    final_time = given_time + timedelta(minutes=n)
    time_intervals.append(final_time)  


time_int = [str(i) for i in time_intervals]
time_int
```

```python
import datetime
timestamp =[]

for i in time_int:
  date1 = datetime.datetime.strptime(i, '%Y-%m-%d %H:%M:%S')
  date_timestamp = datetime.datetime.timestamp(date1)
  timestamp.append(int(date_timestamp))
timestamp
```
